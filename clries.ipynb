```python
import pandas as pd
import matplotlib.pyplot as mlt
import numpy as np
import seaborn as sn
```

```python
d1=pd.read_csv("exercise.csv")
d1.head()
```

```python
d2=pd.read_csv("calories.csv")
d2.head()
```

```python
d=pd.merge(d1,d2,on='User_ID')
d.head()
```

```python
d.info()
```

```python
d.describe()
```

```python
a=pd.get_dummies(d['Gender'],drop_first=True)
a
```

```python
d.head()
```

```python
d=d.drop(['User_ID',"Gender"],axis=1)
```

```python
d.head()
```

```python
d=pd.concat([d,a],axis=1)
d
```

```python
sn.heatmap(d.corr(),annot=True)
```

```python
d.hist(figsize=(20,20),bins=50)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x = d.drop(['Calories'],axis=1)
y = d['Calories']
```

```python
x.head()
```

```python
y.head()
```

```python
xt,xe,yt,ye=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xt = sc.fit_transform(xt)
xe= sc.transform(xe)
```

```python
from sklearn.linear_model import LinearRegression
mod = LinearRegression()
```

```python
mod.fit(xt,yt)
```

```python
mod.intercept_
```

```python
mod.coef_
```

```python
from sklearn.metrics import r2_score
```

```python
ypred=mod.predict(xt)
```

```python
r2_score(yt,ypred)
```

```python
data=pd.DataFrame({"actual":yt,'predict':ypred})
data.head()
```
